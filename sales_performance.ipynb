 IQR method to detect outliers for Quantity, UnitPrice, TotalPrice, Discount, and ShippingCost

```python
columns = ["quantity", "unitprice", "totalprice", "calculated_totalprice","discount", "shippingcost"]

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]

    print(f"\nColumn: {col}")
    print("Lower limit:", lower_limit)
    print("Upper limit:", upper_limit)
    print("Number of outliers:", len(outliers))
```

```python
outlier_rows = pd.DataFrame()

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    temp = df[(df[col] < lower_limit) | (df[col] > upper_limit)].copy()
    temp["Outlier_Column"] = col

    outlier_rows = pd.concat([outlier_rows, temp])

outlier_rows.head()
```

### (iv)CREATING DERIVED COLUMNS


CREATE A COLUMN FOR THE CATEGORY OF THE PRODUCT

```python

# 1. Based on the 'Product' column, define conditions for categorization and create a new 'Category' column.
conditions = [
    df["product"].str.lower().isin(["chair", "desk"]),
    df["product"]
    .str.lower()
    .isin(["laptop", "monitor", "phone", "printer", "tablet"]),
]

choices = ["Furniture", "Electronics"]

# 2. Generate the category values (defaults to 'Other' if a product doesn't match)
category_values = np.select(conditions, choices, default="Other")

# 3. Find the exact position of the 'Region' column and insert 'category' right after it
region_index = df.columns.get_loc("region")
df.insert(loc=region_index + 1, column="category", value=category_values)

# Display the first few rows to verify the position and values
print(df.head())
```

### ️ Feature Engineering & Data Enrichment

This section handles the **Feature Engineering** phase of the data pipeline. By transforming raw transactional fields and dates into structured business metrics, we create the essential **KPIs (Key Performance Indicators)** and **Dimensional Slicers** .

The newly engineered features are organized into four core analytical layers:

### 1. Financial Performance Layer
These metrics track the flow of money from top-line volume down to actual realized revenue:
* **`grosssales`**: Represents the total baseline value of the order before any price cuts ($Quantity \times UnitPrice$).
* **`discountamount`**: Captures the exact monetary value deducted from the gross sales due to promotional discounts ($GrossSales \times Discount$).
* **`netrevenue`**: Reflects the final true financial intake of the business by combining the customer's total payment price with the collected logistics fee (`totalprice` + `shippingcost`).

### 2. Time Intelligence Layer
These features break down exact timestamps into standard financial calendar groupings for time-series trend analysis:
* **`deliverydays`**: Measures operational fulfillment speed by calculating the exact calendar days elapsed between order and delivery.
* **`ordermonth` / `orderquarter` / `orderyear`**: Extracts standardized time periods (e.g., `2026-06`, `2026-Q2`, `2026`) to serve as clean axis coordinates for performance charts.

### 3. Operational & Categorical Layer
Transforms numeric flags and continuous percentages into human-readable categories for descriptive slice-and-dice metrics:

| Engineered Column | Data Type | Transformation Logic | Purpose in Power BI |
| :--- | :--- | :--- | :--- |
| **`returnstatus`** | Text (String) | Converts binary flags (`1`/`0`) into `"Returned"` or `"Not Returned"`. | Direct legend classification for Return Rate charts. |
| **`discountband`** | Categorical | Buckets continuous discount decimals into `"No Discount"`, `"Low"`, `"Medium"`, or `"High"`. | Evaluates promotional campaign impacts on sales volumes. |

---

The Pandas function pd.cut() to transform a continuous numerical column (discount) into a categorical column (DiscountBand). This process is known as data binning or bucketing.This is the raw, numerical source column containing values like 0.0, 0.05, 0.10, or 0.15 (representing 0%, 5%, 10%, and 15% discount rates).The Text Labels: labels=["No Discount", "Low", "Medium", "High"]
This maps friendly text names directly to the 4 buckets we created. The number of labels must exactly match the number of buckets:

Any row falling into the 1st bucket (-0.001 to 0.0) gets the label "No Discount"

Any row falling into the 2nd bucket (0.00 to 0.05) gets the label "Low"

Any row falling into the 3rd bucket (0.05 to 0.10) gets the label "Medium"

Any row falling into the 4th bucket (0.10 to 0.15) gets the label "High"

### (i) Product Frequency Distribution

• ***Visualization Type:-*** Count Plot (Seaborn)  
• ***Purpose:-*** To analyze how frequently each product appears in the order dataset.  
• ***Insight:-*** Products with higher order counts indicate stronger ordering frequency, while lower-count products may require demand improvement or targeted promotion.


Tablets lead with the highest order count, followed closely by laptops and monitors.  
Phones show the lowest order count frequency, indicating stronger customer demand for computing devices compared to accessories or furniture.

```python
# ==============================================================================
# 1. DEFINE CYBER/NEON THEME COLORS (Matching the requested image)
# ==============================================================================
BACKGROUND_COLOR = "#0B0F19"  # Deep dark obsidian blue
BAR_COLOR = "#00E5FF"         # Vibrant neon cyan/light-blue
TEXT_COLOR = "#F3F4F6"        # Clean off-white for high-contrast readability
GRID_COLOR = "#1F2937"        # Muted dark gray for subtle gridlines

# Apply global dark background style
plt.rcParams['figure.facecolor'] = BACKGROUND_COLOR
plt.rcParams['axes.facecolor'] = BACKGROUND_COLOR

# ==============================================================================
# 2. UPDATED POLISH FUNCTION FOR THE DARK THEME
# ==============================================================================
def polish_dark(ax, title, xlabel=None, ylabel=None):
    # Set title with glowing off-white text
    ax.set_title(title, pad=20, color=TEXT_COLOR, fontweight="bold", fontsize=15)
    
    # Configure labels
    if xlabel is not None:
        ax.set_xlabel(xlabel, fontweight="bold", color=TEXT_COLOR, fontsize=11)
    if ylabel is not None:
        ax.set_ylabel(ylabel, fontweight="bold", color=TEXT_COLOR, fontsize=11)
        
    # Style the ticks and axis numbers
    ax.tick_params(colors=TEXT_COLOR, labelsize=10)
    
    # Add subtle background grids to simulate depth
    ax.grid(True, axis='y', color=GRID_COLOR, linestyle='--', linewidth=0.7)
    ax.set_axisbelow(True) # Puts grid lines behind the bars
    
    # Remove the harsh outer border lines (spines) to mimic a floating render
    sns.despine(ax=ax, left=True, bottom=True)
    
    
    return ax
```

```python

# ==============================================================================
# 3. GENERATING THE PLOT
# ==============================================================================
plt.figure(figsize=(5, 3.5))

# Order categories by total counts
order = df["product"].value_counts().index

# Create the plot using the single neon cyber color
ax = sns.countplot(
    data=df, 
    x="product", 
    order=order, 
    color=BAR_COLOR,       # Applies the uniform glowing blue to all bars
    edgecolor="#D4AF37",   # Adds a sharp neon border highlight
    linewidth=1.5,
    alpha=0.9              # Adds a sleek modern opacity look
)

# Apply our dark theme polish
polish_dark(ax, "Product Frequency Distribution", "Product", "Order Count")
plt.xticks(rotation=30, ha="right")

# Add glowing bold data labels directly above the bars
for container in ax.containers:
    ax.bar_label(container, padding=5, fontsize=10, color=BAR_COLOR, fontweight="bold")

plt.tight_layout()
plt.show()
```

### (ii) Payment Method Share

• ***Visualization Type:-*** Donut Chart (Matplotlib)  
• ***Purpose:-*** To analyze how customer orders are distributed across different payment methods.  
• ***Insight:-*** The chart identifies the most commonly used payment methods and helps understand customer payment preferences.


 Payment methods are fairly balanced, with Online (21.5%) slightly leading over Cash and Credit Card. Gift Cards contribute the least (18.4%), suggesting customers prefer direct online or cash-based transactions.


```python


# 1. Count your categories
payment_counts = df["paymentmethod"].value_counts()

# 2. Cyber-Neon configurations
DARK_BG = "#0B0F19"
# Electric cyan, glowing purple, and soft pink/coral for distinct category slices
cyber_colors = ["#00E5FF", "#8B5CF6", "#FF007F","#D4AF37","#57F15F"] 

plt.figure(figsize=(4,4), facecolor=DARK_BG)

# 3. Create the Pie/Donut Slices
wedges, texts, autotexts = plt.pie(
    payment_counts,
    labels=payment_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=cyber_colors,
    pctdistance=0.75,
    # Dark borders separating the wedges look like neon light splits
    wedgeprops={"edgecolor": DARK_BG, "linewidth": 3, "alpha": 0.85} 
)

# 4. Create the center cutout hole (Matching the dark background canvas)
centre_circle = plt.Circle((0, 0), 0.62, fc=DARK_BG)
plt.gca().add_artist(centre_circle)

# 5. Styling the Category Labels and Percentage Values
for t in texts:
    t.set_color("#F3F4F6")      # Glowing off-white outer text
    t.set_fontweight("bold")
    t.set_fontsize(10)
    
for at in autotexts:
    at.set_color("#0B0F19")    # Dark text inside the bright glowing wedges for massive contrast
    at.set_fontweight("bold")
    at.set_fontsize(10)

plt.title("Payment Method Share", fontsize=16, fontweight="bold", color="#F3F4F6", pad=15)
plt.gca().set_facecolor(DARK_BG)
plt.tight_layout()
plt.show()
```

### (iii) Total Price Distribution

• ***Visualization Type:-*** Histogram with KDE Curve (Seaborn)  
• ***Purpose:-*** To understand the distribution and spread of total order values.  
• ***Insight:-*** The chart shows whether most orders are low, medium, or high value and helps identify possible extreme sales values.


The distribution of total prices is right-skewed, with most transactions concentrated between $0–$2,000.Higher-value orders occur less frequently, indicating that sales are driven primarily by lower-priced purchases.


```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

# ==============================================================================
# 1. CYBER/NEON THEME COLORS
# ==============================================================================
BACKGROUND_COLOR = "#0B0F19"  # Deep dark obsidian blue
NEON_CYAN = "#00E5FF"         # Glowing neon light-blue
TEXT_COLOR = "#F3F4F6"        # Off-white for high contrast
GRID_COLOR = "#1F2937"        # Muted grid color

# Set figure and axes background
plt.figure(figsize=(10, 5), facecolor=BACKGROUND_COLOR)
ax = plt.gca()
ax.set_facecolor(BACKGROUND_COLOR)

# ==============================================================================
# 2. GENERATE ADVANCED HISTOGRAM WITH NEON STYLING
# ==============================================================================
# - Replaced 'color' with NEON_CYAN
# - Set 'edgecolor' to match the background so the bins stand out cleanly
# - Tweaked the line_kws to make the KDE line bold and prominent
sns.histplot(
    df["totalprice"], 
    bins=32, 
    kde=True, 
    color=NEON_CYAN, 
    edgecolor=BACKGROUND_COLOR,
    linewidth=1.2,
    alpha=0.4,          # Slight alpha transparency on bars gives a glowing, airy feel
    line_kws={"linewidth": 2.5, "color": NEON_CYAN}  # High-contrast bold curve
)

# ==============================================================================
# 3. APPLY DARK THEME POLISH
# ==============================================================================
ax.set_title("Distribution of Total Price", pad=20, color=TEXT_COLOR, fontweight="bold", fontsize=15)
ax.set_xlabel("Total Price ($)", fontweight="bold", color=TEXT_COLOR, fontsize=11)
ax.set_ylabel("Frequency", fontweight="bold", color=TEXT_COLOR, fontsize=11)

# Configure labels, ticks, and financial formatting
ax.tick_params(colors=TEXT_COLOR, labelsize=10)
ax.xaxis.set_major_formatter(mtick.StrMethodFormatter("${x:,.0f}"))

# Subtle horizontal and vertical grids matching the tech workspace look
ax.grid(True, which='both', color=GRID_COLOR, linestyle='--', linewidth=0.6)
ax.set_axisbelow(True)

# Strip borders for a clean floating canvas
sns.despine(ax=ax, left=True, bottom=True)

plt.tight_layout()
plt.show()
```

## Regional Sales Performance Analysis

This section analyzes sales performance across different regions. It helps identify which regions generate the highest revenue, where customer demand is stronger, and how sales vary geographically.

### (i) Total Sales by Region

• ***Visualization Type:-*** Horizontal Bar Chart (Matplotlib)  
• ***Purpose:-*** To compare total sales generated across different regions.  
• ***Insight:-*** The region with the highest total sales is the strongest contributor to revenue, while lower-performing regions may need targeted marketing or sales improvement strategies.


The North region leads with the highest total sales, followed closely by East and West.  
South shows the lowest performance, highlighting regional disparities and opportunities for targeted growth strategies.

```python

# ==============================================================================
# 1. CYBER/NEON THEME COLORS
# ==============================================================================
BACKGROUND_COLOR = "#0B0F19"  # Deep dark obsidian blue
NEON_CYAN = "#00E5FF"  # Glowing neon light-blue
TEXT_COLOR = "#F3F4F6"  # Off-white for high contrast
GRID_COLOR = "#1F2937"  # Muted grid color

# --- Data Preparation ---
# Note: Changed to ascending=True so the highest sales appear at the TOP of the horizontal bar chart
region_sales = df.groupby("region")["totalprice"].sum().sort_values(ascending=True)

# ==============================================================================
# 2. GENERATE CYBERPUNK HORIZONTAL BAR CHART
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 6), facecolor=BACKGROUND_COLOR)
ax.set_facecolor(BACKGROUND_COLOR)

# Custom color palette with gradient alphas to simulate a glowing look
# The top performing region will have the brightest, most solid color
base_color = sns.color_palette(f"light:{NEON_CYAN}", len(region_sales))

bars = ax.barh(
    region_sales.index,
    region_sales.values,
    color=NEON_CYAN,
    edgecolor=BACKGROUND_COLOR,
    linewidth=1.2,
    alpha=0.85,  # Matches the airy, glowing aesthetic of your histogram
    height=0.65,
)

# Apply subtle color gradations to the bars for visual depth
for idx, bar in enumerate(bars):
    # Progressively brightens bars from bottom to top
    alpha_val = 0.4 + (0.6 * (idx / (len(region_sales) - 1)))
    bar.set_alpha(alpha_val)

# ==============================================================================
# 3. APPLY DARK THEME POLISH
# ==============================================================================
ax.set_title(
    "Total Sales by Region",
    fontsize=18,
    fontweight="bold",
    color=TEXT_COLOR,
    pad=20,
)
ax.set_xlabel("Total Sales ($)", fontweight="bold", color=TEXT_COLOR, labelpad=10)
ax.set_ylabel("Region", fontweight="bold", color=TEXT_COLOR, labelpad=10)

# Configure labels, ticks, and financial formatting
ax.tick_params(colors=TEXT_COLOR, labelsize=11, left=False)
ax.xaxis.set_major_formatter(mtick.StrMethodFormatter("${x:,.0f}"))

# Dynamic text labels over/beside each bar with high contrast styling
for i, value in enumerate(region_sales.values):
    ax.text(
        value,
        i,
        f" ${value:,.0f}",
        va="center",
        ha="left",
        fontsize=11,
        fontweight="bold",
        color=TEXT_COLOR,
    )

# Subtle vertical grids matching the tech workspace look
ax.grid(True, axis="x", color=GRID_COLOR, linestyle="--", linewidth=0.6)
ax.set_axisbelow(True)

# Extend x-axis slightly so labels don't clip off the dark canvas
ax.set_xlim(0, max(region_sales.values) * 1.15)

# Strip borders for a clean floating canvas
sns.despine(ax=ax, left=True, bottom=True)

plt.tight_layout()
plt.show()
```

## Product Revenue Comparison Analysis

This section compares revenue measures across products, especially gross sales and net revenue. It helps understand whether the revenue retained after adjustments differs noticeably across product categories.

### (i) Gross Sales vs Net Revenue by Product

• ***Visualization Type:-*** Grouped Bar Chart (Seaborn / Matplotlib)  
• ***Purpose:-*** To compare gross sales and net revenue for each product after considering discounts and shipping-related values.  
• ***Insight:-*** This chart helps understand how much revenue is retained after adjustments and highlights products where the difference between gross sales and net revenue is more noticeable.


Gross sales consistently exceed net revenue across all products, reflecting the impact of discounts and shipping costs.  
Printers, laptops, and tablets show the strongest revenue contribution, while phones generate the lowest overall returns.


```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import pandas as pd
import seaborn as sns

# ==============================================================================
# 1. CYBER/NEON THEME COLORS
# ==============================================================================
BACKGROUND_COLOR = "#0B0F19"  # Deep dark obsidian blue
NEON_CYAN = "#00E5FF"  # Glowing neon light-blue (Gross Sales)
NEON_PINK = "#f6eedf"  # Vibrant neon pink/magenta (Net Revenue)
TEXT_COLOR = "#F3F4F6"  # Off-white for high contrast
GRID_COLOR = "#1F2937"  # Muted grid color

# --- Data Preparation ---
product_revenue = (
    df.groupby("product")[["grosssales", "netrevenue"]]
    .sum()
    .sort_values("grosssales", ascending=False)
    .reset_index()
)

product_revenue_long = product_revenue.melt(
    id_vars="product",
    value_vars=["grosssales", "netrevenue"],
    var_name="Revenue Type",
    value_name="Amount",
)

# ==============================================================================
# 2. GENERATE CYBERPUNK GROUPED BAR CHART
# ==============================================================================
fig, ax = plt.subplots(figsize=(12, 6), facecolor=BACKGROUND_COLOR)
ax.set_facecolor(BACKGROUND_COLOR)

# Plot using the custom dual neon palette
ax = sns.barplot(
    data=product_revenue_long,
    x="product",
    y="Amount",
    hue="Revenue Type",
    palette=[NEON_CYAN, NEON_PINK],  # Adding a third color for total sales
    edgecolor=BACKGROUND_COLOR,
    linewidth=1.2,
    alpha=0.85,  # Slightly transparent for that airy, digital look
)

# ==============================================================================
# 3. APPLY DARK THEME POLISH
# ==============================================================================
ax.set_title(
    "Gross Sales vs Net Revenue by Product",
    fontsize=18,
    fontweight="bold",
    color=TEXT_COLOR,
    pad=25,
)
ax.set_xlabel("Product", fontweight="bold", color=TEXT_COLOR, labelpad=12, fontsize=12)
ax.set_ylabel("Amount ($)", fontweight="bold", color=TEXT_COLOR, labelpad=12, fontsize=12)

# Format ticks, rotations, and alignment
plt.xticks(rotation=30, ha="right", color=TEXT_COLOR, fontsize=10)
ax.tick_params(axis="y", colors=TEXT_COLOR, labelsize=10)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter("${x:,.0f}"))

# Style the Legend to blend into the cyber deck
legend = ax.legend(
    title="Revenue Type",
    title_fontsize=11,
    fontsize=10,
    facecolor=BACKGROUND_COLOR,
    edgecolor=GRID_COLOR,
)
plt.setp(legend.get_title(), color=TEXT_COLOR, fontweight="bold")
for text in legend.get_texts():
    text.set_color(TEXT_COLOR)

# Dynamic text labels over each bar with proper spacing and coloring
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f"${x:,.0f}" for x in container.datavalues],
        padding=4,
        fontsize=9,
        fontweight="bold",
        color=TEXT_COLOR,
    )

# Subtle horizontal gridlines for data tracking across bars
ax.grid(True, axis="y", color=GRID_COLOR, linestyle="--", linewidth=0.6)
ax.set_axisbelow(True)

# Expand y-limit slightly to prevent top labels from being clipped
ax.set_ylim(0, product_revenue_long["Amount"].max() * 1.15)

# Clean, borderless floating canvas look
sns.despine(ax=ax, left=True, bottom=True)

plt.tight_layout()
plt.show()
```

## Return and Operational Analysis

This section examines returned orders and operational factors. It helps understand which products may create more return-related business risk.

### (i) Return Rate by Product

• ***Visualization Type:-*** Lollipop Chart (Matplotlib)  
• ***Purpose:-*** To identify products with higher return rates.  
• ***Insight:-*** Products with higher return rates may need quality checks, better product descriptions, or improved customer expectation management.


Chairs have the highest return rate (28.2%), followed closely by laptops and monitors.  
Desks show the lowest return rate (21.7%), suggesting product quality or customer satisfaction issues are more concentrated in certain categories.


```python
# ==============================================================================
# 1. CYBER/NEON THEME COLORS
# ==============================================================================
BACKGROUND_COLOR = "#0B0F19"  # Deep dark obsidian blue
NEON_CYAN = "#00E5FF"  # Glowing neon light-blue (Lollipop stems)
NEON_PINK = "#eff669"  # Vibrant neon pink/magenta (Lollipop heads)
TEXT_COLOR = "#F3F4F6"  # Off-white for high contrast
GRID_COLOR = "#1F2937"  # Muted grid color

# --- Data Preparation ---
return_rate = df.groupby("product")["returned"].mean().mul(100).sort_values()

# ==============================================================================
# 2. GENERATE CYBERPUNK LOLLIPOP CHART
# ==============================================================================
fig, ax = plt.subplots(figsize=(10, 4), facecolor=BACKGROUND_COLOR)
ax.set_facecolor(BACKGROUND_COLOR)

# Draw the stems (horizontal lines) using Neon Cyan
ax.hlines(
    y=return_rate.index,
    xmin=0,
    xmax=return_rate.values,
    color=NEON_CYAN,
    linewidth=2.5,
    alpha=0.75,  # Subtle transparency for a glowing stem look
)

# Draw the lollipop heads (markers) using Neon Pink
ax.scatter(
    return_rate.values,
    return_rate.index,
    color=NEON_PINK,
    s=140,  # Slightly enlarged to anchor the visualization
    edgecolor=BACKGROUND_COLOR,  # Dark edge helps it stand out cleanly
    linewidth=1.5,
    zorder=3,  # Ensures the circles sit perfectly on top of the stems
)

# ==============================================================================
# 3. APPLY DARK THEME POLISH
# ==============================================================================
ax.set_title(
    "Return Rate by Product",
    fontsize=18,
    fontweight="bold",
    color=TEXT_COLOR,
    pad=20,
)
ax.set_xlabel("Return Rate (%)", fontweight="bold", color=TEXT_COLOR, labelpad=12)
ax.set_ylabel("Product", fontweight="bold", color=TEXT_COLOR, labelpad=12)

# Configure labels, ticks, and percentage axis formatting
ax.tick_params(colors=TEXT_COLOR, labelsize=11, left=False)
ax.xaxis.set_major_formatter(mtick.PercentFormatter(decimals=0))

# Dynamic text labels placed just to the right of each neon marker
for i, value in enumerate(return_rate.values):
    ax.text(
        value + (max(return_rate.values) * 0.015),  # Proportional padding
        i,
        f"{value:.1f}%",
        va="center",
        ha="left",
        fontsize=11,
        fontweight="bold",
        color=TEXT_COLOR,
    )

# Subtle vertical grids to trace return rates precisely
ax.grid(True, axis="x", color=GRID_COLOR, linestyle="--", linewidth=0.6)
ax.set_axisbelow(True)

# Add right-side buffer spacing so text labels don't get truncated
ax.set_xlim(0, max(return_rate.values) * 1.12)

# Strip borders for a clean floating canvas
sns.despine(ax=ax, left=True, bottom=True)

plt.tight_layout()
plt.show()
```

• ***Purpose:-*** To compare order value spread between customer types.

```python
# ==============================================================================
# 1. CYBER/NEON THEME COLORS
# ==============================================================================
BACKGROUND_COLOR = "#0B0F19"  # Deep dark obsidian blue
NEON_CYAN = "#00E5FF"  # Glowing neon light-blue
NEON_PINK = "#FF007F"  # Vibrant neon pink/magenta
TEXT_COLOR = "#F3F4F6"  # Off-white for high contrast
GRID_COLOR = "#1F2937"  # Muted grid color

# ==============================================================================
# 2. GENERATE CYBERPUNK BOX PLOT
# ==============================================================================
fig, ax = plt.subplots(figsize=(8, 5), facecolor=BACKGROUND_COLOR)
ax.set_facecolor(BACKGROUND_COLOR)

# Custom outlier styling (makes data anomalies look like glowing particles)
flier_props = dict(
    marker="o",
    markerfacecolor="none",
    markeredgecolor=NEON_CYAN,
    markersize=5,
    linestyle="none",
)

# Render the box plot using the dual-neon palette
ax = sns.boxplot(
    data=df,
    x="customertype",
    y="totalprice",
    palette=[NEON_CYAN, NEON_PINK],
    linewidth=1.5,
    fliersize=5,
    flierprops=flier_props,
)

# --- Post-Processing Box Elements for the Cyber Aesthetic ---
# Customizes internal box lines (like the median) to stand out clearly against neon fills
for i, artist in enumerate(ax.artists):
    # Set the bounding box border line to match our background or text color
    artist.set_edgecolor(TEXT_COLOR)

for line in ax.lines:
    # If the line represents the median line inside the box, color it high-contrast off-white
    if line.get_linestyle() == "None" or len(line.get_xdata()) == 2:
        line.set_color(TEXT_COLOR)
        line.set_linewidth(2.0)
    else:
        line.set_color(TEXT_COLOR)

# ==============================================================================
# 3. APPLY DARK THEME POLISH
# ==============================================================================
ax.set_title(
    "Total Price Distribution by Customer Type",
    fontsize=16,
    fontweight="bold",
    color=TEXT_COLOR,
    pad=20,
)
ax.set_xlabel("Customer Type", fontweight="bold", color=TEXT_COLOR, labelpad=12)
ax.set_ylabel("Total Price ($)", fontweight="bold", color=TEXT_COLOR, labelpad=12)

# Configure axis labels and financial tick marks
ax.tick_params(colors=TEXT_COLOR, labelsize=11, bottom=False)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter("${x:,.0f}"))

# Subtle horizontal gridlines across the distribution field
ax.grid(True, axis="y", color=GRID_COLOR, linestyle="--", linewidth=0.6)
ax.set_axisbelow(True)

# Clean, modern borderless aesthetic
sns.despine(ax=ax, left=True, bottom=True)

plt.tight_layout()
plt.show()
```

Bar Chart

**Purpose:**  
To compare the total sales generated by each promotion and identify which promotional offer contributed the most to overall revenue.

**Insights:**  
This chart helps evaluate promotion-wise sales performance. ### (ii) Sales Stacked Bar Chart by Region, Product, and Promotion

• ***Visualization Type:-*** Stacked Bar Chart  (Plotly)  
• ***Purpose:-*** To analyze sales contribution across region, category, and promotion levels together.  
• ***Insight:-*** The chart helps understand how regional sales are distributed across category and promotional categories.


Electronics consistently outperform Furniture across all regions, with the West showing the highest sales contribution.  
Promotions like FREESHIP and WINTER15 drive stronger results compared to SAVE10, especially in Electronics.  
Furniture sales remain modest, highlighting limited promotional impact compared to tech-driven categories.




```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np

# 1. Calculate percentage within each region and category
plot_data = (
    df.groupby(["category", "region", "promotion"])["totalprice"]
    .sum()
    .reset_index()
)
plot_data["percentage"] = (
    plot_data.groupby(["category", "region"])["totalprice"]
    .transform(lambda x: x / x.sum() * 100)
)

# 2. Get unique facets and regions
categories = plot_data["category"].unique()
regions = plot_data["region"].unique()
num_categories = len(categories)

# Setup subplots to mimic the Plotly "facet_col" look
fig, axes = plt.subplots(1, num_categories, figsize=(5 * num_categories, 6), sharey=True)
if num_categories == 1:
    axes = [axes]

# Use a clean qualitative color palette
colors = plt.cm.Set2.colors
promotion_colors = {}

# 3. Build the chart column by column, stack by stack
for ax, cat in zip(axes, categories):
    cat_data = plot_data[plot_data["category"] == cat]
    
    for x_idx, region in enumerate(regions):
        # Isolate data for this specific column, sorted ASCENDING by percentage
        # This guarantees the smallest % is drawn first (bottom) and highest is drawn last (top)
        reg_data = cat_data[cat_data["region"] == region].sort_values("percentage", ascending=True)
        
        bottom_val = 0
        for i, row in reg_data.iterrows():
            promo = row["promotion"]
            perc = row["percentage"]
            
            # Keep color assignment consistent for promotions across the chart
            if promo not in promotion_colors:
                promotion_colors[promo] = colors[len(promotion_colors) % len(colors)]
                
            # Draw the individual segment block
            ax.bar(
                x_idx, perc, bottom=bottom_val, 
                color=promotion_colors[promo], edgecolor='white', width=0.6, label=promo
            )
            
            # Place the percentage label right in the middle of the segment block
            if perc > 3:  # Only label blocks wide enough to fit text
                ax.text(
                    x_idx, bottom_val + (perc / 2), f"{perc:.1f}%", 
                    ha="center", va="center", color="#121212", fontweight="bold", fontsize=9
                )
                
            bottom_val += perc

    # Facet Formatting
    ax.set_title(f"Category: {cat}", fontsize=14, fontweight="bold", pad=12)
    ax.set_xticks(range(len(regions)))
    ax.set_xticklabels(regions, fontsize=11)
    ax.set_xlim(-0.5, len(regions) - 0.5)
    ax.set_ylim(0, 100)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
    ax.grid(axis='y', linestyle=':', alpha=0.5)

axes[0].set_ylabel("Share of TotalSales", fontsize=12, fontweight="bold")

# Deduplicate the legend handles
handles, labels = axes[0].get_legend_handles_labels()
by_label = dict(zip(labels, handles))
fig.legend(by_label.values(), by_label.keys(), title="Promotion", bbox_to_anchor=(1.02, 0.5), loc="center left", fontsize=11)

plt.suptitle("Sales Breakdown by Region & Promotion", fontsize=16, fontweight="bold", y=1.02)
plt.tight_layout()
plt.show()
```
